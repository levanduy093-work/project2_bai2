```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Read and show image from path img/original.jpeg
img = cv2.imread('img/original.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Convert the image to binary image with threshold
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(binary, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
plt.show()
```
